```python
rain = pd.read_csv("/workspaces/CH-dataSience/data/pune_1965_to_2002.csv")
rain = rain.set_index("Year")
rain.sample(10)
x = rain.index
y = rain.sum(axis = 'columns')
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(x, y, label='Precipitaciones acumuladas')
ax.set_xlabel('Ao')                  
ax.set_ylabel(f'Precipitacin acumulada ()')
ax.set_title('Precipitaciones acumuladas segn el ao')
ax.legend() 
```

```python
fig, ax = plt.subplots(figsize=(8,4))
acumulatedRain = rain.sum()
ax.bar(rain.columns, acumulatedRain)
ax.set_title('Precipitaciones acumuladas desde 1965 a 2002, segn el mes')
ax.set_ylabel('Precipitacin total (mm.)')
ax.set_xlabel('Mes')
```

```python
fig, ax = plt.subplots(figsize =  (8,8))
color_mapping = ax.scatter(rain["Nov"], rain["Oct"], c = rain.index)
fig.colorbar(color_mapping)
ax.set_title('Precipitaciones Noviembre-Octubre')
ax.set_xlabel('Precipitaciones en noviembre (mm.)')
ax.set_ylabel('Precipitaciones en octubre (mm.)')
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.boxplot(x=rain.T,)
ax.set_title('Boxplot de precipitaciones')
ax.set_xlabel('Meses')
ax.set_ylabel('Precipitacion (mm)')
```
