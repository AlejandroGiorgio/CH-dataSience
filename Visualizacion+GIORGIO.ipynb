```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Para la segunda entrega se ha elegido trabajar con la dataset de Fraude en solicitud de crédito

```python
df = pd.read_csv("/workspaces/CH-dataSience/data/Base.csv")
df.sample(10)
```

Graficos con MatPlotLib

```python
#histograma para mostrar la frecuencia de la edad de la vitcima
df["customer_age"].plot(kind="hist", bins = 8, figsize= (14,6))
```

```python
#Se aprecia que la mayor parte de las victimas 
#tiene entre 30 y 40 años
```

```python
#grafico de torta para mostrar la distribucion de 
#el sistema operativo de la victima
df["device_os"].value_counts().plot(kind="pie", figsize=(6,6))
```

```python
#se aprecia que macintosh es el OS cuyos usuarios son menos
#vulnerados
#Asi mismo Linux es el OS mas frecuente entre los usuarios
#que son vulnerados
```

```python
#Grafico de densidad para identificar el limite crediticio
#mas usual de la victima
df["proposed_credit_limit"].plot(kind="kde", figsize = (14,6))
```

```python
#Por el contrario de lo que indica el sentido comun
#las personas con un limite crediticio mas bajo (entre 0 a 500) 
#muestran mayor tendencia a ser victimas de una solicitud fraudulenta
```
